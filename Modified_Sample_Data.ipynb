```python
import pandas as pd
```

```python
#Read in select csvs
arrestee = pd.read_csv("NIBRS_ARRESTEE.csv")
incident = pd.read_csv("NIBRS_incident.csv")
offense = pd.read_csv("NIBRS_OFFENSE.csv")
vict = pd.read_csv("NIBRS_VICTIM.csv")
agency = pd.read_csv("agencies.csv")
offense_type = pd.read_csv("NIBRS_OFFENSE_TYPE.csv")
```

```python
#Remove extra columns in each dataframe

arrestee_df = arrestee.drop(['DATA_YEAR','ARRESTEE_SEQ_NUM','AGE_ID','RESIDENT_CODE','UNDER_18_DISPOSITION_CODE',
                                 'CLEARANCE_IND','AGE_RANGE_LOW_NUM','AGE_RANGE_HIGH_NUM'],axis=1)
incident_df = incident.drop(['DATA_YEAR','NIBRS_MONTH_ID','CARGO_THEFT_FLAG','SUBMISSION_DATE',
                             'REPORT_DATE_FLAG','INCIDENT_STATUS','DATA_HOME',
                             'CLEARED_EXCEPT_DATE','ORIG_FORMAT','DID'],axis=1)
vict_df = vict.drop(['DATA_YEAR','ASSIGNMENT_TYPE_ID', 'ACTIVITY_TYPE_ID','OUTSIDE_AGENCY_ID',
                     'AGE_ID','RESIDENT_STATUS_CODE','AGE_RANGE_LOW_NUM','AGE_RANGE_HIGH_NUM'],axis=1)
agency_df = agency[["AGENCY_ID","DATA_YEAR","UCR_AGENCY_NAME","STATE_ID","AGENCY_TYPE_NAME",
                    "POPULATION","SUBURBAN_AREA_FLAG","POPULATION_GROUP_CODE","MALE_OFFICER",
                    "FEMALE_OFFICER","COUNTY_NAME"]]
offense_df = offense.drop(['DATA_YEAR','ATTEMPT_COMPLETE_FLAG','NUM_PREMISES_ENTERED','METHOD_ENTRY_CODE'],axis=1)
```

```python
#Remove records with NaN values
df_lst = [arrestee_df,incident_df,offense_df,vict_df]
for i in df_lst:
    i.dropna(inplace=True)
```

```python
#Merge dataframes
df_1 = pd.merge(incident_df,arrestee_df, on="INCIDENT_ID")
df_2 = pd.merge(df_1,vict_df,on="INCIDENT_ID")
df_3 = pd.merge(df_2,offense_df,on="INCIDENT_ID")
df_4 = pd.merge(df_3,agency_df,on="AGENCY_ID")
df_4.head()
```

```python
df_4.count()
```

```python
Homicide_df = df_4[(df_4["OFFENSE_TYPE_ID_x"].isin([1,32,38]))]
Homicide_df.count()
```

```python
Assault_df = df_4[(df_4["OFFENSE_TYPE_ID_x"].isin([27,44,51]))]
Assault_df.count()
```

```python
HumanTraffick_df = df_4[(df_4["OFFENSE_TYPE_ID_x"].isin([59,60]))]
HumanTraffick_df.count()
```

```python
Kidnapp_df = df_4[(df_4["OFFENSE_TYPE_ID_x"].isin([29]))]
Kidnapp_df.count()
```

```python
SexOffense_df = df_4[(df_4["OFFENSE_TYPE_ID_x"].isin([33,4,36,43,56,3,55]))]
SexOffense_df.count()
```

```python
#Export offense type dataframes to csv's for addtional data analysis

df_names_lst = ["Homicide_df","Assault_df","HumanTraffick_df","Kidnapp_df","SexOffense_df"]
df_lst = [Homicide_df,Assault_df,HumanTraffick_df,Kidnapp_df,SexOffense_df]

df_zip = zip(df_names_lst,df_lst)
for i in df_zip:
    i[1].to_csv(f"{i[0]}.csv",index=False)
```
